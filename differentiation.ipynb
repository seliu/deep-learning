```python
import matplotlib.pyplot as plt
```

```python
def diff( func, x, delta = 1e-4 ):
    return ( func( x + delta ) - func( x ) ) / delta
```

```python
func = lambda x: x**2
x = [ i for i in range( 100 ) ]
```

```python
y = [ func( j ) for j in x ]
fig, ax = plt.subplots()
ax.plot( x, y )
plt.show()
```

```python
y = [ diff( func, j ) for j in x ]
fig, ax = plt.subplots()
ax.plot( x, y )
plt.show()
```
